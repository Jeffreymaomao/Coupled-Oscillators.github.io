### MCMC
Acceptance Rate $x$
$$
\delta = \delta \cdot \underbrace{\bigg(
0.33\left(0.2\tan\left(2.8x-1.26\right)+3.14\right)
\bigg)}_{\text{scale function}}
$$


```python
x1_i = x1_spline(0)
x2_i = x2_spline(0)
v1_i = v1_spline(0)
v2_i = v2_spline(0)
print(f"x1(0) = {x1_i}\nx2(0) = {x2_i} \nv1(0) = {v1_i}\nv2(0) = {v2_i}")
m = (0.7216,0.6950)

def Chi_Prob(t,paras,sigma=1):
    x1_theo, x2_theo, v1_theo ,v2_theo = Analytical_Model(t, k=paras, m=m, x1_i=x1_i, x2_i=x2_i, v1_i=v1_i, v2_i=v2_i)
    x1_data = x1_spline(t)
    x2_data = x2_spline(t)
    # Add 1e-16 prevent overflow
    chi1 = np.abs(np.sum((x1_data - x1_theo)**2 / (x1_theo + 1e-16))) 
    chi2 = np.abs(np.sum((x2_data - x2_theo)**2 / (x2_theo + 1e-16)))
    chi = chi1+chi2
    return exp(-chi/sigma)

def MSE_Prob(t,paras,sigma=1):
    n = len(t)
    x1_theo, x2_theo, v1_theo ,v2_theo = Analytical_Model(t, k=paras, m=m, x1_i=x1_i, x2_i=x2_i, v1_i=v1_i, v2_i=v2_i)
    x1_data = x1_spline(t)
    x2_data = x2_spline(t)
    MSE1 = np.sum((x1_data - x1_theo)**2) /n
    MSE2 = np.sum((x2_data - x2_theo)**2) /n
    MSE = MSE1 + MSE2
    return exp(-MSE/sigma)

Prob = lambda t,paras,sigma=1: MSE_Prob(t,paras,sigma=sigma)

def scale(x):
    if(x<=0 or x>=1.0): return 1.0
    return 0.33*(0.2*np.tan(2.8*x-1.26) + 3.14)

print("Chi_Prob(target) = ", Chi_Prob(t,k_target))
print("MSE_Prob(target) = ", MSE_Prob(t,k_target))
```

```python
print(k_target)

t = np.arange(0, 10, 0.01)
k_ = np.arange(-1,1,0.01) * 400
chi_prob1 = np.zeros(len(k_))
chi_prob2 = np.zeros(len(k_))
chi_prob3 = np.zeros(len(k_))
MSE_prob1 = np.zeros(len(k_))
MSE_prob2 = np.zeros(len(k_))
MSE_prob3 = np.zeros(len(k_))

k_t =  k_target

for i in range(len(k_)):
    chi_prob1[i] = Chi_Prob(t,(k_t[0]+k_[i], k_t[1],       k_t[2])      )
    chi_prob2[i] = Chi_Prob(t,(k_t[0],       k_t[1]+k_[i], k_t[2])      )
    chi_prob3[i] = Chi_Prob(t,(k_t[0],       k_t[1],       k_t[2]+k_[i]))
    MSE_prob1[i] = MSE_Prob(t,(k_t[0]+k_[i], k_t[1],       k_t[2])      )
    MSE_prob2[i] = MSE_Prob(t,(k_t[0],       k_t[1]+k_[i], k_t[2])      )
    MSE_prob3[i] = MSE_Prob(t,(k_t[0],       k_t[1],       k_t[2]+k_[i]))

chi_prob = (chi_prob1,chi_prob2,chi_prob3)
MSE_prob = (MSE_prob1,MSE_prob2,MSE_prob3)

fig,axs = plt.subplots(2,3)
for i in range(3):
    axs[0][i].plot(k_t[i]+k_,chi_prob[i],color='red')
    axs[1][i].plot(k_t[i]+k_,MSE_prob[i],color='blue')
    axs[0][i].set_xlabel(f"$k_{i+1}$")
    axs[1][i].set_xlabel(f"$k_{i+1}$")
    axs[0][i].set_yticks([])
    axs[1][i].set_yticks([])

plt.tight_layout()
plt.subplots_adjust(wspace=0.1)
plt.show()
```

```python
N = 1000
acc_rate = 0
delta = 30.0
# paras = np.array([100.0,100.0,100.0])
paras = np.array(k_target)+1
Paras = np.zeros([N,3])

for i in range(N):
    paras_ = paras + delta * (2*random(3)-1)
    
    p  = Chi_Prob(t, paras)
    p_ = Chi_Prob(t, paras_)
    
    if(p_ >= p or random() < p_/p):
        paras = paras_
        acc_rate += 1
        
    Paras[i] = paras

print("Acceptance Rate : %.5f"%(acc_rate/N))  
    
fig,axs = plt.subplots(3,1,figsize=(12,5), dpi=200)
colors = ["maroon","darkblue","darkgreen"]
for i in range(3):
    axs[i].plot(Paras[:,i], color=colors[i])
    axs[i].set_xticks([])
    axs[i].hlines(y=k_target[i],xmin=0,xmax=N,color='black')
    axs[i].ticklabel_format(useOffset=False)

plt.subplots_adjust(hspace=0.1)
plt.show()
```

```python
t = np.arange(0, 10, 0.01)

N1 = 1000
N2 = 10000
skip = 10
n = 3
Paras  = np.zeros([N2//skip, n])
paras  = np.array([100.0, 100.0, 100.0])

delta  = 40.0
Delta  = np.zeros(N2)
accept = 0
acc_rate = 0.0
Acc_rate = np.zeros(N2)


acc_rate_bar  = FloatSlider(max=1.0,step=0.001)
progress_bar1 = IntProgress(max=N1)
progress_bar2 = IntProgress(max=N2)

print("Acceptance rate")
display(acc_rate_bar)

# Thermal Equilibrium ------------------------------------------------------------------------
print("Thermal Equilibrium")
display(progress_bar1)

for step in range(N1):
    paras_ = paras + delta * (2.0*random(n)-1.0)
    
    p  = Prob(t, paras)
    p_ = Prob(t, paras_)
    if(p_ >= p or random() < p_/p):
        paras = paras_
        accept += 1
    # ---- Acceptance Rate ------------------------
    acc_rate = accept/(step+1)
#     delta = delta*scale(acc_rate)
    # ---- Update Progress Bar------------------------
    acc_rate_bar.value = acc_rate
    progress_bar1.value = step+1
    progress_bar1.description = str((((step+1)*100)//N1))+"%"
    
# Metropolis–Hastings algorithm ------------------------------------------------------------------------
accept = 0
acc_rate = 0.0
delta  = 40.0

print("Metropolis–Hastings algorithm")
display(progress_bar2)

for step in range(N2):
    Delta[step] = delta
    paras_ = paras + delta * (2.0*random(n)-1.0)
    
    p  = Prob(t, paras)
    p_ = Prob(t, paras_)
    if(p_ >= p or random() < p_/p):
        paras = paras_
        accept += 1
    # ---- Save Data ------------------------
    if not step%skip: Paras[step//skip] = paras
    # ---- Acceptance Rate ------------------------
    acc_rate = accept/(step+1)
    Acc_rate[step] = acc_rate
#     delta = delta*scale(acc_rate)
    # ---- Update Progress Bar------------------------
    acc_rate_bar.value = acc_rate
    progress_bar2.value = step+1
    progress_bar2.description = str((((step+1)*100)//N2))+"%"
    
print(f"Acceptance rate: {acc_rate}")
```

```python
fig,axs = plt.subplots(3,1,figsize=(12,5), dpi=200)
axs[0].plot(Paras[:,0], color='maroon')
axs[1].plot(Paras[:,1], color='darkblue')
axs[2].plot(Paras[:,2], color='darkgreen')
[[ax.set_xticks([]),ax.ticklabel_format(useOffset=False)] for ax in axs]
plt.subplots_adjust(hspace=0.1)
plt.show()
```

```python
bins = 100

fig2 = plt.figure(figsize=(8,3), dpi=400)

P1, k1_hist, _ = plt.hist(Paras[:,0], bins=bins, color='maroon', density=True,alpha=0.3,label="$k_1$")
P2, k2_hist, _ = plt.hist(Paras[:,1], bins=bins, color='darkblue', density=True,alpha=0.3,label="$k_2$")
P3, k3_hist, _ = plt.hist(Paras[:,2], bins=bins, color='darkgreen', density=True,alpha=0.3,label="$k_3$")

plt.xlabel("spring constant $k_{1,2,3}$ (N/m)")

plt.tight_layout()
plt.legend(loc='upper right')
plt.show()


k1_max = k1_hist[np.where(P1 == P1.max())][0]
k2_max = k2_hist[np.where(P2 == P2.max())][0]
k3_max = k3_hist[np.where(P3 == P3.max())][0]
print(k1_max,k2_max,k3_max)
print(k_target)
```

```python
error1 = abs(k1_max-k_target[0])/k_target[0]
error2 = abs(k2_max-k_target[1])/k_target[1]
error3 = abs(k3_max-k_target[2])/k_target[2]

error = sqrt(error1**2 + error2**2 + error3**2)
print(f"Error is about {error*100}")
```

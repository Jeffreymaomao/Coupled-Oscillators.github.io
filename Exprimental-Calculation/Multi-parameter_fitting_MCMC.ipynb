```python
#MCMC
def x1(x,k):
    return k[0]*np.sin(k[1]*x)
def x2(x,k):
    return k[0]*np.exp(k[1]*x)


x = np.arange(0, 2*np.pi, 0.01)
k_target = [4, 3.14]
x1_data = x1(x,k_target)
x2_data = x2(x,k_target)

def chi_square(x,k):
    x1_ = x1(x, k)
    x2_ = x2(x, k)
    chi1 = np.abs(np.sum((x1_data - x1_)**2 / (x1_+1e-16)))
    chi2 = np.abs(np.sum((x2_data - x2_)**2 / (x2_+1e-16)))
    return chi1,chi2

N = 100000
n = 2

delta = 1.0
K = np.zeros([N, n])
k = np.array([3.0, 4.0])
for i in range(N):
    k_ = k + delta * (2.0*random(n)-1.0)
    chi1, chi2  = chi_square(x, k)
    chi1_, chi1_ = chi_square(x, k_)
    
    if((chi1_ <= chi1 and chi2_ <= chi2) or (random()  < (chi1*chi2)/(chi1_*chi2_))) :
        k = k_*1.0
    K[i] = k
    
```

```python
plt.figure(figsize=(7,5))
plt.hist(K[:,0], bins=50)
plt.hist(K[:,1], bins=50)
plt.show()
```
